### ADD LETTER NAME AND POSITION TO ALL LETTER IMAGES

```python
import os

# Define the path to the main DATA directory
DATA_DIR = "./DATA"

# Loop through the main directory
for folder in os.listdir(DATA_DIR):
    folder_path = os.path.join(DATA_DIR, folder)

    # Ensure it's a directory
    if os.path.isdir(folder_path):
        for subfolder in os.listdir(folder_path):
            subfolder_path = os.path.join(folder_path, subfolder)

            # Ensure it's a directory
            if os.path.isdir(subfolder_path):
                for file in os.listdir(subfolder_path):
                    file_path = os.path.join(subfolder_path, file)

                    # Check if the file is a .bmp image
                    if os.path.isfile(file_path) and file.endswith(".bmp"):
                        # Get the file name without extension
                        original_name = os.path.splitext(file)[0]

                        # Construct the new name
                        new_name = f"{original_name}_{folder}{subfolder}.bmp"
                        new_file_path = os.path.join(subfolder_path, new_name)

                        # Rename the file
                        os.rename(file_path, new_file_path)

                        print(f"Renamed: {file} -> {new_name}")
```

### ADD _ AFTER THE LETTER NAME AND BEFOR POSITION

```python
import os

# Define the path to the main DATA directory
DATA_DIR = "./DATA1"

# Loop through the main directory
for folder in os.listdir(DATA_DIR):
    folder_path = os.path.join(DATA_DIR, folder)

    # Ensure it's a directory
    if os.path.isdir(folder_path):
        for subfolder in os.listdir(folder_path):
            subfolder_path = os.path.join(folder_path, subfolder)

            # Ensure it's a directory
            if os.path.isdir(subfolder_path):
                for file in os.listdir(subfolder_path):
                    file_path = os.path.join(subfolder_path, file)

                    # Check if the file is a .bmp image
                    if os.path.isfile(file_path) and file.endswith(".bmp"):
                        # Split the file name and extension
                        name, ext = os.path.splitext(file)
                        
                        # Skip if the file already has the correct naming convention
                        if "_" in name.split("_")[-1]:
                            continue
                        
                        # Insert _ before the last letter
                        corrected_name = f"{name[:-1]}_{name[-1]}{ext}"
                        corrected_file_path = os.path.join(subfolder_path, corrected_name)
                        
                        # Rename the file
                        os.rename(file_path, corrected_file_path)

                        print(f"Renamed: {file} -> {corrected_name}")
```

### REMOVE LAST _

```python
import os

# Define the path to the main DATA directory
DATA_DIR = "./DATA1"

# Loop through the main directory
for folder in os.listdir(DATA_DIR):
    folder_path = os.path.join(DATA_DIR, folder)

    # Ensure it's a directory
    if os.path.isdir(folder_path):
        for subfolder in os.listdir(folder_path):
            subfolder_path = os.path.join(folder_path, subfolder)

            # Ensure it's a directory
            if os.path.isdir(subfolder_path):
                for file in os.listdir(subfolder_path):
                    file_path = os.path.join(subfolder_path, file)

                    # Check if the file is a .bmp image
                    if os.path.isfile(file_path) and file.endswith(".bmp"):
                        # Split the file name and extension
                        name, ext = os.path.splitext(file)
                        
                        # Check if the name contains the last `_` before the extension
                        if "_" in name and name[-2] == "_":
                            # Remove the last `_` before the last character
                            original_name = f"{name[:-2]}{name[-1]}{ext}"
                            original_file_path = os.path.join(subfolder_path, original_name)
                            
                            # Rename the file
                            os.rename(file_path, original_file_path)

                            print(f"Renamed: {file} -> {original_name}")
```
